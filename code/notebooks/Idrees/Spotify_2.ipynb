```python
import os
```

```python
files = os.listdir('songs/')
```

```python
files
```

```python
ultra_top = pd.DataFrame()
for file in files:
    data = pd.read_excel(f'songs/{file}')
    data['fileDate'] = f'{file[:8]}'
    ultra_top = ultra_top.append(data)
    
```

```python
ultra_top
```

```python
counting= []
for i in ultra_top.Titre.values:
    counting.append(i)
```

```python
counting
```

```python
from collections import Counter
from typing import List, Dict
Counter(counting)
celeb = dict(Counter(counting))
```

```python
celeb
```

```python
ultra_top['Titre'].map(celeb) 
```

```python
ultra_top
```

```python
ultra_top['freq'] = ultra_top['Titre'].map(celeb) 
```

```python
ultra_top
```
