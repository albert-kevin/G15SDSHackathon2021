## I need to define the predictor variable

```python
import numpy as np
```

```python
def rand_bin_array(K, N):
    arr = np.zeros(N)
    arr[:K]  = 1
    np.random.shuffle(arr)
    return arr

rand_bin_array(5,10)
```

```python
# generatea false set of top 50:
# we need to find a way to correctly populate this list from shazam
df["isInTop50"] = rand_bin_array(50, len(df))
```

### make a report

```python
import pandas_profiling as pp
from IPython.display import Javascript
```

```python
reportFile = "../../../data/report/dataset_fake_target_report.html"
```

```python
# quick report on 100% records (no correlation matrix stuff)
pp.ProfileReport(df=df.sample(frac=1),
                 minimal=True,
                 progress_bar=False,
                 correlations={"cramers": {"calculate": False}}).to_file(reportFile)
```
