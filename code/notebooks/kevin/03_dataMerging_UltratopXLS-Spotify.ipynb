```python
import pandas_profiling as pp
from IPython.display import Javascript
```

```python
reportFile = "../../../data/report/spotify_report.html"
# quick report on 100% records (no correlation matrix stuff)
pp.ProfileReport(df=spotify.sample(frac=1),
                 minimal=True,
                 progress_bar=False,
                 correlations={"cramers": {"calculate": False}}).to_file(reportFile)
```

```python
reportFile = "../../../data/report/ultra_report.html"
# quick report on 100% records (no correlation matrix stuff)
pp.ProfileReport(df=ultra.sample(frac=1),
                 minimal=True,
                 progress_bar=False,
                 correlations={"cramers": {"calculate": False}}).to_file(reportFile)
```
